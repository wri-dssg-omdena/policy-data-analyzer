# Structuring excell sheet info into json

## Imports 

```python
from pathlib import Path
import json
import pandas as pd
import numpy as np
```

## Loading original data form excel

```python
path = Path("../input/")
filename = "WRI Policy Tags Complete File phrases extraction all.xlsx"
file = path / filename
df = pd.concat(pd.read_excel(file, sheet_name = None), ignore_index = True)
```

```python
print(df.shape)
data = df.to_numpy()
df.head(15)
```

```python
dict = {}

i = 0
for item in data:
    if ";" in item[2]:
        sentences = item[2].split(";")
    else:
        i += 1
        print(i, "----", item[2])
        if item[0] in dict:
                dict[item[0]]["Seccion unica"]["sentences"][str(i)] = {"text" : item[2], "labels" : item [4]}
        else:
            dict[item[0]] = {"Seccion unica" : {"tags" : [], "sentences" : {}}}
            dict[item[0]]["Seccion unica"]["sentences"][str(i)] = {"text" : item[2], "labels" : item [4]}
    for sentence in sentences:
        if len(sentence) > 10:
            sentence = sentence.lstrip().rstrip()
            i += 1
            print(i, "----", sentence)#, "\n", item[2])
            if item[0] in dict:
                dict[item[0]]["Seccion unica"]["sentences"][str(i)] = {"text" : sentence, "labels" : item [4]}
            else:
                dict[item[0]] = {"Seccion unica" : {"tags" : [], "sentences" : {}}}
                dict[item[0]]["Seccion unica"]["sentences"][str(i)] = {"text" : sentence, "labels" : item [4]}
    #         print(sentence, "\n***\n")
```

```python
dict
```

```python
sentence_list = []

for key1 in dict:
    for key2 in dict[key1]['Seccion unica']['sentences']:
        text = dict[key1]['Seccion unica']['sentences'][key2]['text']
        labels = dict[key1]['Seccion unica']['sentences'][key2]['labels']
        sentence_list.append([key1, text, labels])
```

```python
path = Path("../output/")
filename = "tagged_sentences_all.csv"
file = path / filename

df = pd.DataFrame(sentence_list)
df.to_csv(file, sep=',',index=False)
```
